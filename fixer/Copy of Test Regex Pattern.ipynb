```python
import numpy as np
import pandas as pd
import random
import string
# from tensorflow.keras.preprocessing.sequence import pad_sequences
# from tensorflow.keras.models import Model
# from tensorflow.keras.layers import Input, LSTM, Dense, Embedding
```

```python
def create_dataset(n_samples=200000):
    names = []
    regex_patterns = []
    pattern_label = []
    # Menambahkan lebih banyak karakter khusus
    special_chars = ['$', '%', '&', '@', '#', '*', '+', '!','|','?']

    for _ in range(n_samples):
        name_length = random.randint(3, 8)
        name = ''.join(random.choices(string.ascii_letters, k=name_length))
        number = random.randint(1, 9999)
        pattern_choice = random.randint(1, 8)  # Diperluas untuk mencakup pola baru
        pattern_label.append(pattern_choice)

        if pattern_choice == 1:
            # Nama diikuti oleh angka
            names.append(f"{name}-{number}")
            regex_patterns.append("^[A-Za-z]+-\d+$")
        elif pattern_choice == 2:
            # Nama dengan angka di tengah
            middle = random.randint(1, name_length-1)
            modified_name = name[:middle] + str(number) + name[middle:]
            names.append(modified_name)
            regex_patterns.append("^[A-Za-z]+\d+[A-Za-z]+$")
        elif pattern_choice == 3:
            # Nama diikuti oleh karakter khusus dan angka
            special_char = random.choice(special_chars)
            names.append(f"{name}{special_char}{number}")
            regex_patterns.append(f"^[A-Za-z]+\{special_char}\d+$")
        elif pattern_choice == 4:
            # Nama diawali angka dan diikuti karakter khusus
            special_char = random.choice(special_chars)
            names.append(f"{number}{special_char}{name}")
            regex_patterns.append(f"^\d+\{special_char}[A-Za-z]+$")
        elif pattern_choice == 5:
            # Hanya huruf
            names.append(name)
            regex_patterns.append("^[A-Za-z]+$")
        elif pattern_choice == 6:
            # Hanya angka
            names.append(str(number))
            regex_patterns.append("^\d+$")
        elif pattern_choice == 7:
            # Format email sederhana
            email = f"{name.lower()}{number}@example.com"
            names.append(email)
            regex_patterns.append("^[a-z0-9]+@[a-z]+\.[a-z]{2,3}$")
        elif pattern_choice == 8:
            # Kombinasi huruf dan karakter khusus (tanpa angka)
            special_char = random.choice(special_chars)
            names.append(f"{name}{special_char}")
            regex_patterns.append(f"^[A-Za-z]+\{special_char}+$")

    return names, regex_patterns, pattern_label
```

```python
names, regex_patterns, pattern_label = create_dataset()
dataset = pd.DataFrame({'name' : names, 'pattern_num' : pattern_label, 'pattern_regex' : regex_patterns})
```

```python
from sklearn.preprocessing import LabelEncoder

encoder           = LabelEncoder()
dataset['label']  = encoder.fit_transform(dataset['pattern_regex'])
label_dict        = dict(zip(encoder.transform(encoder.classes_), encoder.classes_))
dataset.head(50)
```

```python
print(label_dict)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Split data
X_train, X_test, y_train, y_test = train_test_split(dataset['name'], dataset['label'], test_size=0.2, random_state=42)

# Feature extraction
vectorizer = CountVectorizer(analyzer='char', lowercase=False)
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)

# Model training
model_nb = MultinomialNB()
model_nb.fit(X_train_features, y_train)
```

```python
# Predictions
predictions = model_nb.predict(X_test_features)

# Accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Classification report
print("Classification Report:")
print(classification_report(y_test, predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Membuat model Decision Tree
dt_model = DecisionTreeClassifier()

# Melatih model dengan fitur dan label
dt_model.fit(X_train_features, y_train)

# Melakukan prediksi pada set pengujian
dt_predictions = dt_model.predict(X_test_features)

# Melakukan evaluasi
dt_accuracy = accuracy_score(y_test, dt_predictions)
print("Decision Tree Accuracy:", dt_accuracy)

# Classification report
print("Decision Tree Classification Report:")
print(classification_report(y_test, dt_predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Membuat model Random Forest
rf_model = RandomForestClassifier()

# Melatih model dengan fitur dan label
rf_model.fit(X_train_features, y_train)

# Melakukan prediksi pada set pengujian
rf_predictions = rf_model.predict(X_test_features)

# Melakukan evaluasi
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Random Forest Accuracy:", rf_accuracy)

# Classification report
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
```
